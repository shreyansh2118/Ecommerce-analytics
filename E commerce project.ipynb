```python
maxAndMinSales=df.groupby('Order Month')['Sales'].sum().reset_index()
```

# Sales and Profit by Segment

```python
CustomerSeg_sales=df.groupby('Segment')['Sales'].sum().reset_index()
CustomerSeg_sales
```

```python
CustomerSeg_profit=df.groupby('Segment')['Profit'].sum().reset_index()
CustomerSeg_profit
```

```python
total = pd.merge(CustomerSeg_sales, CustomerSeg_profit, on='Segment')
total
```

```python
# Reshaping the data using pd.melt
total_melted = total.melt(id_vars=['Segment'], value_vars=['Sales', 'Profit'], var_name='Metric', value_name='Value')

# Display the reshaped DataFrame
print(total_melted)
```

```python
# Plotting the barplot
plt.figure(figsize=(10, 6))
snc.barplot(
    data=total_melted,
    x='Segment',
    y='Value',
    hue='Metric'  # This will differentiate Sales and Profit by color
)

plt.title('Sales and Profit by Segment')
plt.ylabel('Amount')
plt.show()
```
